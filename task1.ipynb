```python
import pandas as pd
from matplotlib import pyplot as plt
df=pd.read_csv("/content/population.csv")
```

```python
df
```

```python
import matplotlib.pyplot as plt

plot = plt.bar(df['City'], df['Age'])

for bar in plot:
    y = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        y,
        str(y),
        ha='center',

    )

plt.xlabel("City")
plt.ylabel("Age")
plt.title("City-Wise Age Distribution")
plt.show()
```

```python
plot=plt.hist(df['Salary'],color="orange")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.title("Salary wise distribution")
plt.show()
```
